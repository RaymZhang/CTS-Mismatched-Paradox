```python
import numpy as np
import matplotlib.pyplot as plt
from math import floor, ceil
```

```python
t = np.arange(10,1000)


m = 5
l= 1000000


def f(t,m, l):
    return np.sqrt(np.log(t) + (m+2)*np.log(np.log(t)) + m * np.log(1+np.exp(1)/l))/np.sqrt(np.log(t/2) + (m+2)*np.log(np.log(t/2)) + m * np.log(1+np.exp(1)/l))
```

```python
for m in range(1,40,5):
    plt.plot(t,f(t,m,l), label = f"{m}")


plt.legend()
```
