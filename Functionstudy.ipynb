```python
import numpy as np
import matplotlib.pyplot as plt
from math import floor, ceil

from scipy import optimize 
```

# Lemma F1

```python
x = np.linspace(3, 100000, 1000)

def f(x):
    return np.log(np.log(x))*np.log(x/2) - np.log(x)*np.log(np.log(x/2)) 


def fprime(x):
    return 1/(x*np.log(x))*np.log(x/2) + np.log(np.log(x))/x - np.log(np.log(x/2))/x - 1/(x*np.log(x/2))*np.log(x)
```

```python
plt.figure("fonction1",figsize=(16, 10))
plt.plot(x, f(x), label="f(x)")
plt.plot(x, np.zeros(len(x)) )
plt.legend()
```

```python
root = optimize.newton(f, 5, fprime = fprime, tol=1e-8, maxiter=10000)
```

```python
print(root)
```
